```python
# -*- coding: utf-8 -*-
"""
Created on Tue Aug 18 10:03:33 2020

@author: chen
"""

import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return math.sin(x/10.0+0.2)+0.1
    
def f2(x):
    return math.acos(math.sin(x/8.0))/2.1

def f3(x):
    return math.sin(x/6.0)*math.cos(x/12.0)*1.3-0.12
    
def f4(x):
    return math.sin(x/10.0-0.01)*math.cos(x/10.0+0.55)**2+0.1

def f5(x):
    return math.sin(x/9.5)**2*math.cos(x/9.5-0.1)*1.6

def f0(x):
    return f1(x)+f2(x)+f3(x)+f4(x)+f5(x)
```

```python
if __name__ == "__main__":
    l=300
    listx=[i for i in range(1,l)]
    list1=[f1(i) for i in range(1,l)]
    list2=[f2(i) for i in range(1,l)]
    list3=[f3(i) for i in range(1,l)]
    list4=[f4(i) for i in range(1,l)]
    list5=[f5(i) for i in range(1,l)]
    list0=[f0(i) for i in range(1,l)]
    
    plt.figure()
    plt.subplot(611)
    plt.plot(listx,list1)
    plt.subplot(612)
    plt.plot(listx,list2)
    plt.subplot(613)
    plt.plot(listx,list3)
    plt.subplot(614)
    plt.plot(listx,list4)
    plt.subplot(615)
    plt.plot(listx,list5)
    plt.subplot(616)
    plt.plot(listx,list0)
    plt.show()
```
